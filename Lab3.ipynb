```python
#import pandas to Google Colab

import pandas as pd
```

```python
#using a very big dataset,
#download California datasets file from Github
```

```python
#load California training datasets to Google Colab

df_train=pd.read_csv("/content/california_housing_train.csv")

df_train
```

```python
#do the same for testing datasets

df_test=pd.read_csv("/content/california_housing_test.csv")

df_test
```

```python
#define data to use for training and testing.
```

```python
#define for x_train and y_train

x_train=df_train.loc[:,['housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income']]

y_train=df_train.loc[:,['median_house_value']]

#x_train
y_train
```

```python
#do the same for x_test and y_test

x_test=df_test.loc[:,['housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income']]

y_test=df_test.loc[:,['median_house_value']]

#x_test
y_test
```

```python
#Implementing Decision Tree Regression to evaluate datasets
```

```python
#Decision Tree Regression

#train DT model

from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(x_train, y_train)
```

```python
#create a model to predict test data

y_predict=model.predict(x_test)

y_predict
```

```python
#using evaluation metrics to test the model

from sklearn import metrics
import numpy as np

print('Mean Squared Error: ', metrics.mean_squared_error(y_test, y_predict))
print('Mean Absolute Error: ', metrics.mean_absolute_error(y_test, y_predict))
print('Root Mean Squared Error: ', np.sqrt(metrics.mean_squared_error(y_test, y_predict)))
```

```python
#Implementing Random Forest Regression to evaluate datasets
```

```python
#Random Forest Regression

#train Randon Forest model

from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators = 200)
model.fit(x_train, y_train)
```

```python
#create a model to predict test data

y_predict=model.predict(x_test)

y_predict
```

```python
#using evaluation metrics to test the model

print('Mean Squared Root: ', metrics.mean_squared_error(y_test, y_predict))
print('Mean Absolute Error: ', metrics.mean_absolute_error(y_test, y_predict))
print('Root Mean Squared Error: ', np.sqrt(metrics.root_mean_squared_error(y_test, y_predict)))
```

```python
#Implementing Linear Regression to evaluate datasets
```

```python
#Linear Regression

#train Linear Regression model

from sklearn.linear_model import LinearRegression
model_LR=LinearRegression()
model_LR.fit(x_train, y_train)
```

```python
#create a model to predict test data

y_predict=model_LR.predict(x_test)

y_predict
```

```python
#using evaluation metrics to test the model

print('Mean Squared Error: ', metrics.mean_squared_error(y_test, y_predict))
print('Mean Absolute Error: ', metrics.mean_absolute_error(y_test, y_predict))
print('Root Mean Squared Error: ', np.sqrt(metrics.root_mean_squared_error(y_test, y_predict)))
```

```python
#Implementing Support Vector Regression to evaluate datasets
```

```python
#Support Vector Regression

#Train SVR model

from sklearn.svm import SVR
model = SVR(kernel = 'linear')
model.fit(x_train, y_train)
```

```python
#create a model to predict test data

y_predict=model.predict(x_test)

y_predict
```

```python
#using evaluation metrics to test the model

print('Mean Squared Error: ', metrics.mean_squared_error(y_test, y_predict))
print('Mean Absolute Error: ', metrics.mean_absolute_error(y_test, y_predict))
print('Root Mean Squared Error: ', np.sqrt(metrics.root_mean_squared_error(y_test, y_predict)))
```
