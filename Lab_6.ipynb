```python
#importing libraries needed to perform Clustering

import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
#Using IRIS Datasets,
#upload dataset to Google Colab
#df means DataFrame

df = pd.read_csv('/content/IRIS.csv')

df
```

```python
#Select features to use from IRIS Datasets
#Meaning we want to choose column to use
```

```python
#Choose petal_length and petal_width as features to use.
#Plot scatter chart of petal_length vs petal_width

plt.scatter(df['petal_length'], df['petal_width'])

plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.title('Scatter Plot of Petal Length and Width');


```

```python
#Using KMeans Clustering
```

```python
#Define df1 from IRIS DataFrame

df1= df[['petal_length','petal_width']]

df1
```

```python
#Define cluster with KMeans

#(n_cluster=3) means we're specifying that the algorithm should,
#find 3 clusters in the Scatter Chart of Petal Length and Width.

km=KMeans(n_clusters=3, random_state=0)
```

```python
#Here we're predicting the clusters of each data labels

#km.labels

predict_y = km.fit_predict(df1)

predict_y
```

```python
#km.cluster_centers_ gives the centriods of each cluster.
#Each row represents the mean (petal_length,	petal_width) for a cluster.
#km.cluster_centers_[0] gives the centriod (mean) of the first cluster.

km.cluster_centers_
```

```python
#Add the cluster lables to the original IRIS DataFrame as a new column 'cluster'

df['cluster']=predict_y

df
```

```python
#Define each cluster. Since cluster is [0,1,2]

df_c0 = df[df.cluster==0]
df_c1 = df[df.cluster==1]
df_c2 = df[df.cluster==2]
```

```python
#Plot the data points for each cluster with different colors

plt.scatter(df_c0['petal_length'], df_c0['petal_width'], label='Cluster 0', color='red')
plt.scatter(df_c1['petal_length'], df_c1['petal_width'], label='Cluster 1', color='green')
plt.scatter(df_c2['petal_length'], df_c2['petal_width'], label='Cluster 2', color='blue')

#Adding (x and y label) & legend.

plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.legend();
```

```python
#Using Scaling,
#We can calculate the min & max values of our data
#Scaling always range [0, 1]
```

```python
#Cluster are formed properly, scaling is required

scaler= MinMaxScaler()
scaler.fit(df[['petal_width']])
df[['petal_width']]= scaler.transform(df[['petal_width']])

df
```

```python
#Do scaling for petal_length

scaler.fit(df[['petal_length']])
df[['petal_length']]= scaler.transform(df[['petal_length']])

df
```

```python
#use KMeans for the new DataFrame,

km= KMeans(n_clusters=3, random_state=0)
```

```python
#After scaling for both petal_length & petal_width
#Display only the columns

df1= df[['petal_length','petal_width']]

#df1
```

```python
#Here predicting the new clusters of each data labels

predict_y = km.fit_predict(df1)

predict_y

```

```python
#Add the cluster lables to the original IRIS DataFrame as a new column 'New_clusters'

df['New_clusters']= predict_y

df
```

```python
#Define each new cluster. Since new cluster is [0,1,2]

df_c0 = df[df.New_clusters==0]
df_c1 = df[df.New_clusters==1]
df_c2 = df[df.New_clusters==2]

#Plot the data points for each cluster with different colors

plt.scatter(df_c0['petal_length'], df_c0['petal_width'], label='Cluster 0', color='purple')
plt.scatter(df_c1['petal_length'], df_c1['petal_width'], label='Cluster 1', color='green')
plt.scatter(df_c2['petal_length'], df_c2['petal_width'], label='Cluster 2', color='blue')

#Adding (x and y labels) & legend.

plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.legend();
```

```python
#km.cluster_centers_ gives the centriods of each new cluster.
#Each row represents the mean (petal_length,	petal_width) for a new cluster.
#km.cluster_centers_[0] gives the centriod (mean) of the first cluster.

km.cluster_centers_
```

```python
#Plotting the new cluster and showing the centriods with a marker.

df_c0 = df[df.New_clusters==0]
df_c1 = df[df.New_clusters==1]
df_c2 = df[df.New_clusters==2]

plt.scatter(df_c0['petal_length'], df_c0['petal_width'], label='Cluster 0', color='purple')
plt.scatter(df_c1['petal_length'], df_c1['petal_width'], label='Cluster 1', color='green')
plt.scatter(df_c2['petal_length'], df_c2['petal_width'], label='Cluster 2', color='blue')

plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1],color='red', marker='*', label='Centroid')

plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.legend();
```

```python
#Using Elbow Method,
#Calculate Sum of Square Error(SSE)

K_range= range(1,10)
sse=[]

for k in K_range:
  km= KMeans(n_clusters=k)
  km.fit(df[['petal_length','petal_width']])
  sse.append(km.inertia_)

sse
```

```python
#Plot Elbow Chart

plt.xlabel('K')
plt.ylabel('Sum of Square Error')
plt.plot(K_range, sse)

plt.title('Elbow Chart');

```
