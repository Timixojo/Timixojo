```python
#since it has alraedy been installed;

#Import matplotlip and Seborn

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Using Line Chart to visualise data
```

```python
#Here is a Python list showing the yield of apples (tons per hectare) over six years in an imaginary country called Wuse.

yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
```

```python
#visualize how the yield of apples changes over time using a line chart.
#To draw a line chart, we can use the plt.plot function.

plt.plot(yield_apples)
```

```python
#I do not want the output to show [<matplotlib.lines.Line2D at 0x7d7f40b831d0>]
#only display the graph as output.
#Semicolon will be use with plt.plot

plt.plot(yield_apples);
```

```python
#What if there is an actual years for X_axis instead of using (0-5)
#Then, we will declare those years for X_axis

years = [2010, 2011, 2012, 2013, 2014, 2015]
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]

plt.plot(years, yield_apples);


```

```python
#What is A graph without proper labels.
#plt.xlabel and plt.ylabel methods.

plt.xlabel("Years")
plt.ylabel("Yield (tons per hectare)")
plt.plot(years, yield_apples);

```

```python
#Plotting Multiple Line Graph
#Declaring the years and produce
#Here we are using range for the years

years = range(2010, 2016)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
oranges = [0.962, 0.941, 0.93, 0.923, 0.918, 0.908]

plt.xlabel("Years")
plt.ylabel("Yield (tons per hectare)")

plt.plot(years, apples)
plt.plot(years, oranges);
```

```python
#Graph needs Title and Legend for better visualisation
#Here both plt.title and plt.legend will be used

plt.xlabel("Years")
plt.ylabel("Yield (tons per hectare)")

plt.title("Crop Yield in Wuse Over 5 Period of Years")

plt.plot(years, apples)
plt.plot(years, oranges)

plt.legend(["Apples", "Oranges"]);

```
